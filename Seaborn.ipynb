```python
from matplotlib import pyplot
from sklearn import datasets
import seaborn as sns
from pandas import DataFrame

#чтобы графики выводились сразу же, без plt.plot()
%pylab inline  
```

```python
iris = datasets.load_iris()
iris_frame = DataFrame(iris.data)

#Имена признаков
print('feature names: {}'.format(iris.feature_names))

#Имена классов
print('feature names: {}'.format(iris.target_names))

iris_frame.columns = iris.feature_names # в качестве названия колонок - имена признаков
iris_frame['target'] = iris.target # чтобы анализировать данные, зная информацию о классах

#Заменим метки классов на их имена (с помощью функции apply)
#lambda-функция берёт метку класса и вместо неё подставляет название (0 -> имя (setosa))

iris_frame.target = iris_frame.target.apply(lambda x: iris.target_names[x])
#Посмотрим на набор данных:
iris_frame.head()
```

```python
#в hue говорим, в каком столбике находится целевая метка
sns.pairplot(iris_frame, hue='target')
```
