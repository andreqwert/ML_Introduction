```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold, cross_val_score
```

Требуется предсказать возраст ракушки по физическим измерениям

```python
# Загрузите данные
data = pd.read_csv('abalone.csv')
```

```python
data.head()
```

```python
# Преобразуйте признак Sex в числовой
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
```

```python
data.head()
```

```python
# Разделите содержимое файла .csv на признаки и целевую переменную.
# В последнем столбце записана целевая переменная, в остальных - признаки

data_target = data.Rings.values # отрежем целевую переменную

# Отдельно отрежем данные
data_features = data.iloc[:, :-1]

y = data_target
X = data_features
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=1)
```

```python
# Обучите случайный лес с различным числом деревьев от 1 до 50

def test_accuracy(cv, X, y):
    for k in range(1, 51):
        rf = RandomForestRegressor(n_estimators=k, random_state=1)
        rf.fit(X, y)
        score = cross_val_score(rf, X, y, cv=cv, scoring="r2").mean()
        print('Num of trees = {};'.format(k), 'score = {}'.format(score))

test_accuracy(cv, X, y)
```
