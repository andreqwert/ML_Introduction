Рассмотрена сложная математическая функция на отрезке. В этом скрипте приближена эта функция многочленами при помощи решения системы линейных уравнений.

Аппроксимация проходит по точкам: 1, 4, 10, 15. Затем: 1, 16, 100, 225 Потом: $1*1$, $4*16$, $10*100$, $15*225$.

# Функция, которую будем аппроксимировать

```python
# Рассмотрим сложную математическую функцию на отрезке [1, 15]:
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
    return y


x = np.arange(1, 16) # функция на отрезке [1, 15]
y = f(x)
plt.xlabel('$x$') 
plt.ylabel('$f(x)$')
plt.plot(x, y)
plt.grid()
plt.show()
```

# Аппроксимация полиномом первого порядка
Многочлен первой степени сопадает в точках 1 и 15

```python
# Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A и свободный вектор b) 
# для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15. 
# Решите данную систему с помощью функции scipy.linalg.solve. 


a = np.array([[1., 1.], [1., 15.]]) # матрица коэффициентов 
b = np.array([f(1.), f(15.)]) # свободный вектор
xpol1 = np.linalg.solve(a, b) # ответ при полиноме первой степени
print('Значения аппроксимирующей функции: ', xpol1)

def f1(x):
    return xpol1[0] + xpol1[1] * x

x = np.arange(1, 16)
y1 = f1(x)
plt.plot(x, y1)
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$') 
plt.grid(True)
plt.show()
```

# Аппроксимация полиномом второго порядка
Многочлен первой степени сопадает в точках 1, 8 и 15

```python
a = np.array([[1., 1., 1.], [1., 8., 64.], [1., 15., 225.]])
b = np.array([f(1.), f(8.), f(15.)])
xpol2 = np.linalg.solve(a, b)
print('Значения аппроксимирующей функции: ', xpol2)

def f2(x):
    return xpol2[0] + xpol2[1] * x + xpol2[2] * x ** 2

x = np.arange(1, 16) # функция на отрезке [1, 15]
y2 = f2(x)
plt.xlabel('$x$') 
plt.ylabel('$f(x)$')
plt.plot(x, y2)
plt.grid()
plt.show()
```

# Аппроксимация полиномом третьего порядка
Многочлен первой степени сопадает в точках 1, 4, 10 и 15

```python
a = np.array([[1., 1., 1., 1.], [1., 4., 16., 64.], [1., 10., 100., 1000.], [1., 15., 225., 3375.]])
b = np.array([f(1.), f(4.), f(10.), f(15.)])
xpol3 = np.linalg.solve(a, b)
print('Значения аппроксимирующей функции: ', xpol3)

def f3(x):
    return xpol3[0] + xpol3[1] * x + xpol3[2] * x ** 2 + xpol3[3] * x ** 3


x = np.arange(1, 16) # функция на отрезке [1, 15]
y3 = f3(x)
plt.xlabel('$x$') 
plt.ylabel('$f(x)$')
plt.plot(x, y3)
plt.grid()
plt.show()
```
